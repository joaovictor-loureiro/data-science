# Panorama da Covid-19 no Brasil e no mundo

## A origem do coronavírus

**08 de dezembro de 2019**, data em que os primeiros casos de **coronavírus (Covid-19)** foram registrados em um hospital na cidade de **Wuhan**, capital da província de Hubei, na China. Um mercado de frutos do mar, que também comercializava animais silvestres, é apontado como o local de origem do contágio, uma vez que, os primeiros contaminados eram frequentadores dessa localidade.

A principal hipótese diz que, provavelmente, o vírus foi transmitido de morcegos para um outro animal intermediário para depois chegar nos humanos. Esse animal intermediário ainda é desconhecido pelos pesquisadores. Além do morcego, outros artigos científicos também relataram a presença do vírus em pangolins, animais que são usados tradicionalmente na medicina chinesa. Ainda que, em princípio, não fossem vendidos no mercado de Wuhan, não é possível descartar a possibiladade de venda ilegal.

Os sintomas da infecção pelo coronavírus podem variar de um simples resfriado, a uma síndrome gripal, até uma pneumonia severa. Sendo tosse, febre, coriza, dor de garganta, dificuldade respiratória, perda de olfato, alteração no paladar, distúrbios gastrintestinais (vômitos e diarréia), cansaço e falta de ar, os sintomas mais comuns.

Algumas pessoas têm maior chance de desenvolver as formas mais graves da doença, e essas fazem parte do chamado grupo de risco, que são os portadores de doenças crônicas, idosos acima de 60 anos, grávidas, entre outros.
Porém, é importante destacar também que, a maioria dos infectados não apresentam sintomas do Covid-19. O que torna a doença ainda mais contagiosa, visto que, um assintomático pode transmitir para outras pessoas sem saber que está fazendo isso.

A transmissão acontece de uma pessoa infectada para outra, ou por contato por meio de gotículas de saliva, espirro, tosse e objetos ou superfícies contaminadas. O diagnóstico da doença é feito através da técnica de RT-PCR com uma amostra colhida da orofaringe (parte da garganta atrás da boca), sendo que todo paciente com suspeita de infecção pelo coronavírus deve ter seu caso notificado para o órgão sanitário responsável, e o paciente deverá se manter em isolamento domiciliar.

No **dia 11 de março de 2020**, o diretor-geral da Organização Mundial da Saúde (OMS), Tedros Adhanom anunciou que a Covid-19, seria agora caracterizada como uma **pandemia**. E recomendou que todos os países adotassem estratégias de prevenção, para salvar vidas e minimizar o impacto em seus habitantes.

Ainda de acordo com a OMS, pandemia é um termo usado para uma determinada doença que rapidamente se espalhou por diversas regiões do mundo através de uma contaminação sustentada. Nesse quesito, a gravidade da doença não é determinante e sim o seu poder de contágio e sua proliferação geográfica.

Hoje (28 de março de 2021), cerca de 381 dias após o início da pandemia, serão mostrados neste notebook dados e métricas do impacto da Covid-19 na população de alguns países. O objetivo desta análise é tentar mensurar e exibir de forma simples as dimensões dessa pandemia que o mundo vem enfrentando.

### Visualização dos dados

Para conseguir visualizar melhor as informações que esse dataframe trás, serão plotados gráficos de cada uma das variáveis numéricas presentes na base de dados.

#### Parâmetros e funções para exibição dos gráficos

A seguir serão criadas duas variáveis que representam a data em que foi feita a última atualização dos dados, uma variável será representada no formato DIA/MÊS/ANO e a outra no formato ANO-MÊS-DIA, que é o formato presente no dataframe. Isso irá auxiliar em trazer os dados mais recentes para os gráficos.

```
# Data da última atualização dos dados
last_update = str(df['date'].tail(1))
last_update = last_update.split('\n')
last_update = last_update[0]
last_update = last_update.split('    ')
last_update = last_update[1]
last_update = last_update.split('-')
last_update_format_day = last_update[2] + '/' + last_update[1] + '/' + last_update[0]

# Data da última atualização dos dados (formato YYYY-MM-DD)
last_update_format_year = last_update[0] + '-' + last_update[1] + '-' + last_update[2]

# Deixando este notebook com informações estáticas, do dia 28/03/2021.
# Caso queira as informações atuais, basta apagar as duas linhas seguintes
last_update_format_day = '28/03/2021' 
last_update_format_year = '2021-03-28'

# Imprimindo a data de atualização dos dados
print(last_update_format_day)
print(last_update_format_year)
```

Também será criada uma função para formatar os números que serão exibidos nos gráficos, para deixar a visualização dos mesmos mais agradável.

```
# Criando uma função para formatar os números
def formatarString(value):
  value = int(value)
  value = str(value)

  if(len(value) == 4):
    value = value[:1] + '.' + value[1:]
  elif(len(value) == 5):
    value = value[:2] + '.' + value[2:]
  elif(len(value) == 6):
    value = value[:3] + '.' + value[3:]
  elif(len(value) == 7):
    value = value[:1] + '.' + value[1:4] + '.' + value[4:]
  elif(len(value) == 8):
    value = value[:2] + '.' + value[2:5] + '.' + value[5:]
  elif(len(value) == 9):
    value = value[:3] + '.' + value[3:6] + '.' + value[6:]

  return value
```

Além disso, serão criados dicionários para armazenar os dados que serão exibidos nos gráficos.

```
# Criando um dicionário para a porcentagem de pessoas contaminadas em cada continente, ao redor do mundo
world_cases = int(df.loc[df['date'] == last_update_format_year, 'total_cases'].sum())
continent_cases = df.loc[df['date'] == last_update_format_year, ['continent', 'total_cases']]
continent_cases['total_cases'] = (continent_cases['total_cases']).fillna(0)
continent_cases['total_cases'] = (continent_cases['total_cases']).astype(int)
continent_cases = continent_cases.groupby('continent')['total_cases'].sum().sort_values(ascending=False)
dict_continents_cases = dict(round((continent_cases / world_cases) * 100, 2))

# Criando uma variável apenas para a taxa de pessoas contaminadas na Oceania, pois o continente possui uma taxa
# muito baixa e não será exibida no gráfico. E transformando ela em uma string para exibir nas informações.
oceania_cases = str(dict_continents_cases['Oceania'])
oceania_cases = oceania_cases.replace('.', ',')

# Removendo a Oceania do dicionário principal, pois seus dados não serão exibidos
dict_continents_cases.pop('Oceania')

# Exibindo como ficou o dicionário principal, da taxa de contaminados em cada continente. E a taxa da Oceania.
print(dict_continents_cases)
print(" 'Oceania': {}".format(oceania_cases))
```

```
# Criando um dicionário com os países com maior número de casos de Covid-19
countries_more_cases = df.loc[df['date'] == last_update_format_year, ['location', 'total_cases']].sort_values(by='total_cases', ascending=False).head(10).copy()
countries_more_cases['location'] = countries_more_cases['location'].str.lower()
countries_more_cases['location'] = countries_more_cases['location'].str.replace(' ', '-')
countries_more_cases.set_index('location', inplace=True)
countries_more_cases = dict(countries_more_cases['total_cases'].astype(int))

# Criando um dicionário com os países com maior número de mortes por Covid-19
countries_more_deaths = df.loc[df['date'] == last_update_format_year, ['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False).head(10).copy()
countries_more_deaths['location'] = countries_more_deaths['location'].str.lower()
countries_more_deaths['location'] = countries_more_deaths['location'].str.replace(' ', '-')
countries_more_deaths.set_index('location', inplace=True)
countries_more_deaths = dict(countries_more_deaths['total_deaths'].astype(int))

# Criando um dicionário com os países que mais aplicaram doses de vacina para o Covid-19
countries_more_vaccinations = df.loc[df['date'] == last_update_format_year, ['location', 'total_vaccinations']].sort_values(by='total_vaccinations', ascending=False).head(10)
countries_more_vaccinations['location'] = countries_more_vaccinations['location'].str.lower()
countries_more_vaccinations['location'] = countries_more_vaccinations['location'].str.replace(' ', '-')
countries_more_vaccinations.set_index('location', inplace=True)
countries_more_vaccinations = countries_more_vaccinations['total_vaccinations'].astype(int)
countries_more_vaccinations = countries_more_vaccinations.to_dict()
```

E por fim, será criado um dicionário que armazenará o ícone das bandeiras de alguns países que irão aparecer nos gráficos. O intuito é deixar as informações mais claras sobre cada país e trazer uma boa ilustração para os gráficos.

```
# Pegando as listas de todos os países que irão aparecer nos gráficos
list(countries_more_cases.keys())
list(countries_more_deaths.keys())
list(countries_more_vaccinations.keys())

# Criando uma lista para guardar todos esses países
all_countries_flags = []

for country in countries_more_cases.keys():
  all_countries_flags.append(country)

for country in countries_more_deaths.keys():
  if country in all_countries_flags:
    country = country
  else:
    all_countries_flags.append(country)

for country in countries_more_vaccinations.keys():
  if country in all_countries_flags:
    country = country
  else:
    all_countries_flags.append(country)

# Criando um dicionário com os links do GitHub, para adicionar o ícone de cada país aos gráficos
flags = {}

for i in range(len(all_countries_flags)):
  flags[all_countries_flags[i]] = 'https://raw.githubusercontent.com/joaovictor-loureiro/data-science/master/data-science/arquivos/imagens/flags/' + all_countries_flags[i] + '.png'

flags['erro'] = 'https://raw.githubusercontent.com/joaovictor-loureiro/data-science/master/data-science/arquivos/imagens/flags/error.png';

# Definindo o tamanho dos ícones
zoom = 0.055
```

#### O impacto da Covid-19 em escala continental

```
# Gráfico do número de casos da Covid-19 distribuídos em cada continente
fig, ax = plt.subplots(figsize=(14, 8))

ax.barh(tuple(dict_continents_cases.keys()), tuple(dict_continents_cases.values()), color=['#f2ebeb'])
ax.get_xaxis().set_ticks([])
ax.invert_yaxis()
ax.set_xlim(xmin=0.0, xmax=60)
ax.set_facecolor("#fff")
ax.set_title('\nO IMPACTO DA COVID-19 EM ESCALA CONTINENTAL\n', color='#bdb7b7', horizontalalignment='right', fontsize=22)
ax.spines['left'].set_color('#bdb7b7')
ax.tick_params(axis='y', colors='#ad9a9a', labelsize=20)

ax.annotate(formatarString(world_cases), xy=(45, -0.02), fontsize=45, fontweight='bold', color='#BA0B0B', ha='center')
ax.annotate("pessoas contaminadas", xy=(45, 0.35), fontsize=27, fontweight='normal', color='#D46A6A', ha='center')

y = .085
for continent in dict_continents_cases.values():
  ax.annotate('{}%'.format(round(continent)), xy=(continent - .75, y), fontsize=20, fontweight='normal', color='#ad9a9a', ha='right')
  y += 1

# Criando anotações no gráfico
ax.annotate('% de pessoas contaminadas por continente. Dados de {}.'.format(last_update_format_day), xy=(41.7, 4.64), fontsize=12, fontweight='normal', color='#ad9a9a', ha='center')
ax.annotate('A Oceania não está presente \nno gráfico pois sua taxa \nrepresenta cerca de {}% \ndo total de contaminados.'.format(oceania_cases)
          , xy=(45, 3.24), fontsize=20, fontweight='normal', color='#60b0b3', ha='center', linespacing=1.5)

plt.tight_layout()

# Exportando o gráfico
fig.savefig('o-impacto-da-covid-19-em-escala-continental.png', bbox_inches='tight', dpi=300)
```

Informações que podem ser identificadas:

* Em 28 de março de 2021, mais de um ano após o início da pandemia, quase **128 milhões de pessoas** já foram infectadas pelo Covid-19.
* A Europa é o quarto continente mais populoso do mundo, porém é o continente que lidera o ranking de maior número de pessoas contaminadas. Cerca de **31% do total de contaminados** em todo o mundo, **são habitantes do continente europeu**.
* América do Norte, Ásia e América do Sul representam, respectivamente, 27%, 22% e 16% do total de contaminados.
* O continente africano, apesar de ser o segundo mais populoso, possui uma taxa de contaminação relativamente baixa. Com cerca de 3% do total de contaminados.
* E a **Oceania**, continente formado por várias ilhas do Oceano Pacífico e com o menor número de habitantes, possui uma taxa de contaminados de apenas **0,03%**. Taxa tão baixa, que sua barra nem chega a ser representada no gráfico.

#### Países com os maiores números de contaminados pela Covid-19

```
# Parâmetros para as cores das barras do gráfico
i = 0
color = []
for country in countries_more_cases.keys():
  if country == 'brazil':
    color.append('#BA0B0B')
  else:
    color.append('#f2ebeb')

# Gráfico dos países com maior número de casos
fig, ax = plt.subplots(figsize=(14, 8))

ax.barh(tuple(countries_more_cases.keys()), tuple(countries_more_cases.values()), color=color)
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ax.invert_yaxis()
ax.set_xlim(xmin=-2000000, xmax=list(countries_more_cases.values())[0])
ax.set_facecolor("#fff")
ax.set_title('\nPaíses com os maiores números de contaminados pela Covid-19\n', color='#bdb7b7', horizontalalignment='center', fontsize=22)
ax.tick_params(axis='y', colors='#ad9a9a', labelsize=20)

# Parâmetros para os ícones
xbox = -1200000
ybox = .04

# Exibindo ícones de cada país
for country in countries_more_cases.keys():
  try:
    flag = mpimg.imread(flags[country])
  except:
    flag = mpimg.imread(flags['erro'])
  flag = OffsetImage(flag, zoom=zoom, resample=True)
  flag = AnnotationBbox(flag, (0, 0), frameon=False, xybox=(xbox, ybox))
  ybox += 1
  ax.add_artist(flag)

# Parâmetros para os números dos 3 países com mais casos
i = 0 
ybox = .15
xbox = 3*1000000

# Exibindo os números dos 3 países com mais casos
while(i < 3):
  top_cases = formatarString(str(list(countries_more_cases.values())[i]))
  if(list(countries_more_cases.keys())[i] == 'brazil'):
    color = '#fff'
  else:
    color = '#BA0B0B'
  ax.annotate(top_cases, xy=(list(countries_more_cases.values())[i] - xbox, ybox), fontsize=20, fontweight='bold', color=color, ha='center')
  i += 1 
  ybox += 1


# Criando anotações no gráfico
ax.annotate('O Brasil é o segundo país \ncom o maior número de \ninfecções pela Covid-19. \nSão mais de {} milhões \nde pessoas contaminadas.'.format(str(countries_more_cases['brazil'])[:2]), 
            xy=(list(countries_more_cases.values())[0], 6.24), fontsize=19, fontweight='normal', color='#ad9a9a', ha='right', linespacing=1.5)
ax.annotate('Número de pessoas contaminadas em cada país. Dados atualizados em {}.'.format(last_update_format_day), 
            xy=(list(countries_more_cases.values())[0], 9.83), fontsize=12, fontweight='normal', color='#ad9a9a', ha='right')

plt.tight_layout()

# Exportando o gráfico
fig.savefig('paises-com-maiores-numeros-de-contaminados.png', bbox_inches='tight', dpi=300)
```

Informações que podem ser identificadas:

* **Estados Unidos é o país com o maior número de contaminados** pela Covid-19, cerca de **30 milhões de pessoas** já foram infectadas.
* O Brasil é o segundo país com mais contaminados, **mais de 12 milhões de brasileiros** já entraram para essa estatística. 
* A Índia, que é o segundo país mais populoso do mundo, aparece como terceiro país com o maior número de pessoas contaminadas pelo coronavírus.
* Entre os países com os maiores números de pessoas contaminadas, também aparecem (em ordem decrescente): França, Rússia, Reino Unido, Itália, Espanha, Turquia e Alemanha.

#### Países com os maiores números de mortos pela Covid-19

```
# Parâmetros para as cores das barras do gráfico
color = []
for country in countries_more_cases.keys():
  if country == 'brazil':
    color.append('#BA0B0B')
  else:
    color.append('#f2ebeb')

# Gráfico dos países com maior número de mortes por Covid
fig, ax = plt.subplots(figsize=(14, 8))

ax.barh(tuple(countries_more_deaths.keys()), tuple(countries_more_deaths.values()), color=color)
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ax.invert_yaxis()
# ax.set_xlim(xmin=0.0, xmax=60)
ax.set_facecolor("#fff")
ax.spines['left'].set_color('#d9d9d9')
ax.set_title('\nPaíses com os maiores números de mortos pela Covid-19\n', color='#bdb7b7', horizontalalignment='right', fontsize=22)
ax.tick_params(axis='y', colors='#ad9a9a', labelsize=20)

# Parâmetros para os ícones
xbox = -22500
ybox = .04

# Exibindo ícones de cada país
for country in countries_more_deaths.keys():
  try:
    flag = mpimg.imread(flags[country])
  except:
    flag = mpimg.imread(flags['erro'])
  flag = OffsetImage(flag, zoom=zoom, resample=True)
  flag = AnnotationBbox(flag, (0, 0), frameon=False, xybox=(xbox, ybox))
  ybox += 1
  ax.add_artist(flag)

# Parâmetros para exibir os números de mortes
ybox = .15

# Exibindo o número de mortes de cada país
for value in countries_more_deaths.values():
  if ybox > 3:
    color = '#bdb7b7'
  elif value == countries_more_deaths['brazil']:
    color = '#FFF'
  else:
    color = '#BA0B0B'
  ax.annotate(formatarString(value), xy=(value - 5000, ybox), fontsize=20, fontweight='bold', color=color, ha='right')
  ybox += 1

# Criando anotações no gráfico
ax.annotate('O Brasil é o segundo país \ncom o maior número de \nmortes pela Covid-19. \nSão mais de {} mil pessoas \nque vieram a óbito.'.format(str(countries_more_deaths['brazil'])[:3]), 
            xy=(list(countries_more_deaths.values())[0], 6.24), fontsize=19, fontweight='normal', color='#ad9a9a', ha='right', linespacing=1.5)
ax.annotate('Número de mortes por Covid-19 em cada país. Dados atualizados em {}.'.format(last_update_format_day), 
            xy=(list(countries_more_deaths.values())[0], 9.83), fontsize=12, fontweight='normal', color='#ad9a9a', ha='right')

plt.subplots_adjust(left=0.01, right=0.99, top=0.8, bottom=0.2)

plt.tight_layout()

# Exportando o gráfico
fig.savefig('paises-com-maiores-numeros-de-mortos.png', bbox_inches='tight', dpi=300)
```

Isso reforça a ideia de que quanto maior for a contaminação dessa doença, maior também será sua letalidade na população.

#### O impacto da Covid-19 no Brasil

No Brasil, o primeiro caso de Covid-19 foi confirmado no dia 26 de fevereiro de 2020. E algumas semanas depois, no dia 17 de março de 2020 foi notificada a primeira morte por coronavírus no Brasil. Desde então começaram as primeiras ações governamentais tentando frear a propagação do vírus e conter o impacto da pandemia na vida dos brasileiros. Infelizmente, tais medidas não foram suficientes ou tão eficazes como em outros países, e como visto nos gráficos acima, hoje, após um ano do início da pandemia, o **Brasil registra 12 milhões de contaminados e mais de 300 mil mortos**.

Para tentar entender como o país chegou a esses números, nos gráficos abaixo serão mostrados os números de novas contaminações e novas mortes a cada mês, desde o início da pandemia, no Brasil.

##### Número de novos casos mensais de Covid-19 no Brasil

```
# Criando um dataframe com os dados exclusivos do Brasil
data_Brazil = df.loc[df['location'] == 'Brazil', ['date', 'location', 'new_cases', 'new_deaths']]
data_Brazil.set_index('date', inplace=True)

# Agrupando o número de novos casos de acordo com cada mês
data_Brazil_cases = pd.DataFrame(data_Brazil.groupby(pd.to_datetime(data_Brazil.index).strftime('%Y/%m'))['new_cases'].sum().sort_index())

# Transformando os números em tipo inteiro
data_Brazil_cases['new_cases'] = data_Brazil_cases['new_cases'].astype(int)

# Pegando apenas os valores de novos casos mensais
casos_mensais_Brazil = dict(data_Brazil_cases['new_cases'])

# Parâmetros para as cores das barras do gráfico
color = []
for month in casos_mensais_Brazil.keys():
  if casos_mensais_Brazil[month] == casos_mensais_Brazil[max(casos_mensais_Brazil, key=casos_mensais_Brazil.get)]:
    color.append('#BA0B0B')
  else:
    color.append('#f2ebeb')

# Plotando o Gráfico de novos casos de Covid em cada mês, no Brasil
fig, ax = plt.subplots(figsize=(11, 9))

data_Brazil.groupby(pd.to_datetime(data_Brazil.index).strftime('%Y/%m'))['new_cases'].sum().sort_index().plot(kind='barh', width=.75, color=color)
ax.set_title('\nNúmero de novos casos mensais de Covid-19 no Brasil\n', color='#bdb7b7', horizontalalignment='center', fontsize=22)
ax.set_ylabel('')
ax.set_facecolor("#fff")
ax.get_xaxis().set_ticks([])
ax.tick_params(axis='y', colors='#ad9a9a', labelsize=12)
ax.invert_yaxis()

# Parâmetro para exibir os números de casos
ybox = .17

# Exibindo o número de casos no Brasil
for month in casos_mensais_Brazil.keys():
  if ybox > 3:
    if casos_mensais_Brazil[month] == casos_mensais_Brazil[max(casos_mensais_Brazil, key=casos_mensais_Brazil.get)]:
      color = '#FFF'
    else:
      color = '#bdb7b7'    
    ax.annotate(formatarString(casos_mensais_Brazil[month]), xy=(casos_mensais_Brazil[month] - 10000, ybox), fontsize=18, fontweight='bold', color=color, ha='right')  
  ybox += 1


# Criando anotações no gráfico
ax.annotate('Dados atualizados em {}.'.format(last_update_format_day), xy=(casos_mensais_Brazil[max(casos_mensais_Brazil, key=casos_mensais_Brazil.get)], -0.5), 
            fontsize=12, fontweight='normal', color='#ad9a9a', ha='right')

plt.tight_layout()

# Exportando o gráfico
fig.savefig('casos-mensais-de-covid-19-Brasil.png', bbox_inches='tight', dpi=300)
```

Informações que podem ser identificadas:

* **Março de 2021** vem sendo o mês em que mais casos de Covid-19 foram registrados no Brasil, foram quase **2 milhões de novos casos**. É o **pior mês de contaminação** desde o início da pandemia.
* Também é possível observar que após atingir mais de 1 milhão de pessoas contaminadas por mês, nos meses de Julho e Agosto de 2020, o número de novos contaminados começou a diminuir. Porém, em Novembro, esse número voltou a aumentar, batendo o recorde de pior mês em Janeiro de 2021, e quebrando esse recorde em Março. </br>
Esses números talvez estejam relacionados com alguns fatores que podem ter contribuído para esse grande aumento na contaminação, em Novembro de 2020 foram realizadas as eleições municipais em todo o país, o que pode ter causado aglomerações tanto na campanha de alguns candidatos como no dia da votação.</br>
Além disso, em dezembro houve flexibilização nas regras de isolamento e inúmeras festas de fim de ano, e os infectologistas já alertavam que isso era um perigo para a proliferação do vírus. Portanto, esse pico de novos casos em Março de 2021 pode ser um reflexo dos meses anteriores.

```
# Agrupando o número de novas mortes de acordo com cada mês
data_Brazil_cases = pd.DataFrame(data_Brazil.groupby(pd.to_datetime(data_Brazil.index).strftime('%Y/%m'))['new_deaths'].sum().sort_index())

# Transformando os números em tipo inteiro
data_Brazil_cases['new_deaths'] = data_Brazil_cases['new_deaths'].astype(int)

# Pegando apenas os valores de novos casos mensais
mortes_mensais_Brazil = dict(data_Brazil_cases['new_deaths'])

# Parâmetros para as cores das barras do gráfico
color = []
for month in casos_mensais_Brazil.keys():
  if casos_mensais_Brazil[month] == casos_mensais_Brazil[max(casos_mensais_Brazil, key=casos_mensais_Brazil.get)]:
    color.append('#BA0B0B')
  else:
    color.append('#f2ebeb')

# Plotando o Gráfico de novas mortes por Covid em cada mês, no Brasil
fig, ax = plt.subplots(figsize=(11, 9))

data_Brazil.groupby(pd.to_datetime(data_Brazil.index).strftime('%Y/%m'))['new_deaths'].sum().sort_index().plot(kind='barh', width=.75, color=['#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#f2ebeb','#BA0B0B',])
ax.set_title('\nNúmero de novas mortes mensais pela Covid-19 no Brasil\n', color='#bdb7b7', horizontalalignment='center', fontsize=20)
ax.set_ylabel('')
ax.set_facecolor("#fff")
ax.get_xaxis().set_ticks([])
ax.tick_params(axis='y', colors='#ad9a9a', labelsize=12)
ax.invert_yaxis()

# Parâmetro para exibir os números de casos
ybox = .17

# Exibindo o número de casos no Brasil
for month in mortes_mensais_Brazil.keys():
  if ybox > 2:
    if mortes_mensais_Brazil[month] == mortes_mensais_Brazil[max(mortes_mensais_Brazil, key=mortes_mensais_Brazil.get)]:
      color = '#FFF'
    else:
      color = '#bdb7b7'    
    ax.annotate(formatarString(mortes_mensais_Brazil[month]), xy=(mortes_mensais_Brazil[month] - 500, ybox), fontsize=18, fontweight='bold', color=color, ha='right')  
  ybox += 1


# Criando anotações no gráfico
ax.annotate('Dados atualizados em {}.'.format(last_update_format_day), xy=(mortes_mensais_Brazil[max(mortes_mensais_Brazil, key=mortes_mensais_Brazil.get)], -0.5), 
            fontsize=12, fontweight='normal', color='#ad9a9a', ha='right')

plt.tight_layout()

# Exportando o gráfico
fig.savefig('mortes-mensais-de-covid-19-Brasil.png', bbox_inches='tight', dpi=300)
```

Informações que podem ser identificadas:

* É possível perceber que esse gráfico segue a mesma tendência do gráfico anterior, do número de contaminações mensais, e como já dito anteriormente, era esperado que quanto maior a contaminação, maior seria a letalidade da doença. </br>
É visível que esse alto número de mortos no mês de março está associado ao gigantesco número de pessoas contaminadas no mesmo mês. Inúmeros hospitais estão em colapso, e em muitos não há mais leitos de UTI (Unidade de Terapia Intensiva) disponíveis. </br>
Um boletim do Observatório Covid-19 da Fiocruz (Fundação Oswaldo Cruz), divulgado no dia 23 de março de 2021, diz que o colapso do sistema de saúde em praticamente todo o país torna o coronavírus ainda mais mortal.</br>
A Fiocruz ainda aponta que *\"este colapso não foi produzido em março de 2021, mas ao longo de vários meses, refletindo os modos de organização para o enfrentamento da pandemia no país, nos Estados e nos municípios”*.

#### Vacinação contra a Covid-19

Ao longo de 2020, a comunidade científica do mundo todo se voltou para tentar desenvolver uma vacina que conseguisse combater o coronavírus. E após muito esforço dos pesquisadores e cientistas, os resultados começaram a aparecer. De acordo com a OMS, em fevereiro de 2021 pelo menos 7 imunizantes já haviam sido lançados pelo mundo. Atualmente, há 13 vacinas aprovadas e 83 em fase de testes clínicos. E segundo especialistas, esse número deve continuar subindo.

O objetivo das vacinas é diminuir a gravidade da infecção e, consequentemente, a mortalidade causada pela doença. As novas vacinas contra a Covid-19 estão se mostrando capazes disso.  Esse fato traz um certo alívio às pessoas, que podem começar a imaginar um fim para essa pandemia após a imunização do maior número de pessoas possíveis.

No gráfico abaixo, serão mostrados os países que conseguiram vacinar o maior número de pessoas até agora.


```
# Parâmetros para as cores das barras do gráfico
color = []
i = 0
for country in countries_more_vaccinations.keys():
  if i < 3:
    color.append('#744EEE')
  else:
    color.append('#EAE4FE')
  i += 1

# Vacinação
fig, ax = plt.subplots(figsize=(14, 8))

ax.barh(tuple(countries_more_vaccinations.keys()), tuple(countries_more_vaccinations.values()), color=color)
ax.invert_yaxis()
ax.set_facecolor("#fff")
ax.spines['left'].set_color('#d9d9d9')
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ax.set_title('\nPaíses com os maiores números de vacinados\n', color='#ad9a9a', horizontalalignment='right', fontsize=22)
ax.tick_params(axis='y', colors='#ad9a9a', labelsize=20)

# Parâmetros para os ícones
xbox = -5000000
ybox = .04

# Exibindo ícones de cada país
for country in countries_more_vaccinations.keys():
  try:
    flag = mpimg.imread(flags[country])
  except:
    flag = mpimg.imread(flags['erro'])
  flag = OffsetImage(flag, zoom=zoom, resample=True)
  flag = AnnotationBbox(flag, (0, 0), frameon=False, xybox=(xbox, ybox))
  ybox += 1
  ax.add_artist(flag)

plt.tight_layout()

# Parâmetros para os números dos 3 países com mais pessoas vacinadas
i = 0 
ybox = .15
xbox = 13*1000000

# Exibindo os números dos 3 países com mais pessoas vacinadas
while(i < 3):
  top_cases = formatarString(str(list(countries_more_vaccinations.values())[i]))
  ax.annotate(top_cases, xy=(list(countries_more_vaccinations.values())[i] - xbox, ybox), fontsize=20, fontweight='bold', color='#fff', ha='center')
  i += 1 
  ybox += 1

# Criando anotações no gráfico
ax.annotate('{}'.format(formatarString(countries_more_vaccinations['brazil'])), 
            xy=(list(countries_more_vaccinations.values())[0], 5), fontsize=50, fontweight='bold', color='#744EEE', ha='right', linespacing=1.5)
ax.annotate('pessoas vacinadas no Brasil', xy=(list(countries_more_vaccinations.values())[0], 5.7), fontsize=22, fontweight='normal', color='#b8aeae', ha='right', linespacing=1.5)
ax.annotate('Número de pessoas vacinadas em cada país. Dados atualizados em {}.'.format(last_update_format_day), 
            xy=(list(countries_more_vaccinations.values())[0], 9.83), fontsize=12, fontweight='normal', color='#ad9a9a', ha='right')

plt.tight_layout()

# Exportando o gráfico
fig.savefig('paises-com-maiores-numeros-vacinados.png', bbox_inches='tight', dpi=300)
```

* China e Índia vêm logo em seguida, com 106 e 60 milhões de pessoas vacinadas, respectivamente.
* O **Brasil** aparece na quarta posição dessa lista, com mais de **17 milhões de pessoas vacinadas**.

## Conclusão

É notório que ninguém esperava que uma doença misteriosa, que surgiu na China no final de 2019, iria tomar as proporções que tomou. Mais de um ano já se passou, e o mundo ainda segue enfrentando a Covid-19. Alguns países talvez já tenham passado pela sua pior fase da pandemia, porém outros, ainda sofrem com o impacto do coronavírus.

Mais de 128 milhões de pessoas já foram infectadas pelo Covid-19. E o continente europeu, sozinho, representa 31% desse número. Já os países da Oceania, representam um número incrivelmente baixo de pessoas contaminadas, apenas 0,03% do total de infectados são de lá.

Estados Unidos é o país com o maior número de pessoas contaminadas, são mais de 30 milhões. Brasil e Índia vêm em seguida, com cerca de 12 milhões de contaminados cada um.

O país com maior número de mortos também é os Estados Unidos, mais de 500 mil pessoas já vieram a óbito por conta da Covid-19. Brasil aparece como segundo novamente, com mais de 300 mil brasileiros mortos.

No Brasil, a pandemia ainda segue descontrolada. E segundo especialistas, essa foi uma tragédia anunciada.

E por fim, após meses angustiantes de espera, finalmente as pessoas ao redor do mundo começaram a ser vacinadas. Atualmente já existem 13 vacinas aprovadas, e os Estados Unidos já conseguiram vacinar mais de 140 milhões de pessoas. No Brasil, o número de vacinados ultrapassou os 17 milhões.

Agora resta esperar que essa pandemia chegue ao fim logo, e que a Covid-19 não passe de um triste momento histórico.
