# Matplotlib para visualização de dados

**Matplotlib** é uma biblioteca para visualização de dados do Python. Ela disponibiliza diversas ferramentas para construção de gráficos e para facilitar a visualização dos dados.

Neste notebook irei mostrar alguns exemplos práticos da utilização da biblioteca **Matplotlib** para geração de gráficos.

##Obtenção dos dados

Afim de demonstrar na prática alguns exemplos utilizando o Matplotlib, será importado um *dataset* da plataforma **Kaggle**.

O *dataset* em questão se trata de uma compilação de informações sobre os municípios brasileiros (segue o [link](https://www.kaggle.com/crisparada/brazilian-cities), em caso de interesse).

##Bibliotecas necessárias

Para conseguir utilizar a biblioteca **Matplotlib** é necessário importa-lá no notebook.

Além dela, também será importada a biblioteca **Pandas** para manusear os dados do *dataset*.

```
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
```

##Importando os dados

Importando os dados do *dataset* "brazilian-cities", contendo informações sobre todas as cidades brasileiras.

```
# Importando o dataset
database = pd.read_csv('https://raw.githubusercontent.com/joaovictor-loureiro/data-science/master/data-science/datasets/BRAZIL_CITIES.csv', sep=';')
```

```
# Ver as primeiras entradas
database.head()
```

##Gráfico de Barras

O gráfico de barras é composto por barras de formato retangular cujo tamanho é proporcional ao valor que ela representa.

O gráfico abaixo, feito utilizando a biblioteca **Matplotlib** mostra os 5 estados brasileiros com maior número de municípios, trazendo uma visualização bastante clara sobre esses números.

```
# Plotando o gráfico
database['STATE'].value_counts().head(5).plot(kind='barh', figsize=(10, 5), color=['#539460', '#61ad70', '#6ec47f', '#7ee091', '#8cfaa1'], zorder=2, width=0.75)

plt.title('Estados com as maiores quantidades de municípios')
plt.xlabel('Quantidade de municípios')
plt.ylabel('Estado')
plt.xticks(rotation=0)
plt.show()
```

O próximo gráfico de barras trás informações sobre as cidades brasileiras com maior número de habitantes, sendo duas delas localizadas na região Sudeste do Brasil. Veja:

```
# Pegando apenas os dados que serão utilizados no gráfico
maiores_cidades = database[['CITY','IBGE_RES_POP']].sort_values(by='IBGE_RES_POP', ascending=False).head(5)
maiores_cidades = maiores_cidades.set_index('CITY')

# Plotando o gráfico
maiores_cidades.plot(kind='barh', figsize=(10, 5), color='#66d17b', zorder=2, width=0.75, legend=None)
plt.xticks(rotation=0)
plt.title('As cidades brasileiras com maior número de habitantes')
plt.xlabel('Número de habitantes')
plt.ylabel('Cidade')
plt.ticklabel_format(style='sci', axis='x', scilimits=(1,10))
plt.show()
```

Por fim, este último gráfico de barras mostra informações sobre as cidades brasileiras com maior número de habitantes estrangeiros, ou seja, pessoas de outros países que agora estão morando no Brasil.

```
# Pegando apenas os dados que serão utilizados no gráfico
cidades_mais_estrangeiros = database[['CITY', 'IBGE_RES_POP_ESTR']].sort_values(by='IBGE_RES_POP_ESTR', ascending=False).head(5)
cidades_mais_estrangeiros = cidades_mais_estrangeiros.set_index('CITY')

# Plotando o gráfico
cidades_mais_estrangeiros.plot(kind='barh', figsize=(10, 5), color='#eb6363', zorder=2, width=0.75, legend=None)
plt.xticks(rotation=0)
plt.title('As cidades brasileiras com maior número de habitantes estrangeiros')
plt.xlabel('Número de habitantes estrangeiros')
plt.ylabel('Cidade')
plt.show()
```

##Gráfico de Dispersão

Um outro tipo de gráfico, que também pode ser plotado utilizando a biblioteca **Matplotlib**, é o de dispersão (ou Scatter Plot).

Se trata de um gráfico bidimensional para visualização de dados, que usa pontos para representar o valor entre duas variáveis diferentes.

No exemplo abaixo, usa-se os dados de longitude e latitude de cada cidade para plotar uma visualização geral da distribuição dos municípios no território brasileiro. Veja:

```
# Plotando o gráfico
database.plot.scatter(x='LONG', y='LAT', s=0.1, figsize=(10, 10), color='#3040f0')
plt.show()
```

Esses foram alguns exemplos de gráficos plotados utilizando a biblioteca **Matplotlib**, além do Gráfico de Barras e do Gráfico de Dispersão, a biblioteca fornece outros inúmeros gráficos para as mais diferentes situações.
